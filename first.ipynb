```python
#starting opencv

import cv2
```

```python
img=cv2.imread("DATA/00-puppy.jpg")
```

```python
plt.imshow(img[:,:,2],cmap="gray")
```

```python
plt.imshow(img[:,:,0],cmap="gray")
```

```python
plt.imshow(img)
```

```python
#MATPLOT-RED,GREE,BLUE
#OPENCV-BLUE,GREEN,RED
```

```python
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
#HOW TO CONVERT A IMAGE INTO GRAY SCALE IMAGE IN OPENCV
img_gray=cv2.imread("DATA/00-puppy",cv2.IMREAD_GRAYSCALE)
```

```python
type(img_gray)
```

```python
plt.imshow(fix_img)
```

```python
new_img=cv2.resize(fix_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
##CHANGING THE IMAGE BY RATIOS OF WIDTH AND HEIGHT

```

```python
w_ratio=0.5
h_ratio=.5
```

```python
new_img=cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
#HOW TO FLIP AN IMAGE WITH HORIZIMTAL-0,VERTICAL-1,BOTH=-1
```

```python
new_img=cv2.flip(fix_img,0)
```

```python
plt.imshow(new_img)
```

```python
#HOW TO SAVE AN IMAGE FILE WHICH IS NUMPYARRAY
```

```python
cv2.imwrite("totallyneimage.jpg",new_img)
```

```python
p=cv2.imread("totallyneimage.jpg")
```

```python
plt.imshow(p)
```
